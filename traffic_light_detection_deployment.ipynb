```python
from ultralytics import YOLO
```

```python
onnx_model = YOLO("/content/drive/MyDrive/yolo_traffic_project/best.onnx")
```

```python
import gradio as gr
import cv2

def greet(video):
    cap = cv2.VideoCapture(video)
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    output_path = '/content/drive/MyDrive/yolo_traffic_project/output_video.mp4'
    out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        #cv2.imshow('frame', frame)
        results = onnx_model(frame)
        annotaed_img=results[0].plot()
        out.write(annotaed_img)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    # print(video)
    processed_video = gr.Video(output_path)
    return processed_video

demo = gr.Interface(fn=greet, inputs="video", outputs="video")

demo.launch(share=True,debug=True)  # Share your demo with just 1 extra parameter 🚀
```
